# lec 48 : Distribution plots

```python
import seaborn as sns
import scipy.stats as stats
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(15)
```

```python
sns.set_style("whitegrid") # sns.set_style('ticks') else  it can be
# darkgrid, whitegrid, dark, white, and ticks
```

```python
sns.distplot(tips['total_bill'])
sns.despine()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30) 
# distplot is for univarient data 
```

```python
# joint plot is for bi-varient data

sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')

# kind could be reg,hex,default is scatter plot, kde
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.set_style("darkgrid")
sns.pairplot(tips,hue="sex")
```

```python
# kde plots = Kernel Density Estimation plots

sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

but there are different ways whcih we can tranform a dataset into a

```python
 fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

# lec 51 : Grids

```python
 import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
```

```python
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
g.map(sns.distplot,'total_bill')
```

```python
g.map(sns.histplot,'total_bill')
```

```python
g.map(sns.displot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```
